```python
plt.xlabel('Total Rooms')
plt.ylabel('Row Count')
plt.title('Histogram of Total Rooms')
data['total_rooms'].hist(figsize=(10,5),color="indigo",ec="black",lw=3) #hist(figsize=(x,y),color="copper",ec="black",lw=3))

#x axis = total room value
#y axis = row count
```

```python
plt.xlabel('Total bedrooms')
plt.ylabel('Row Count')
plt.title('Histogram of Total bedrooms')
data['total_bedrooms'].hist(figsize=(10,5),color="cyan",ec="black",lw=3) #hist(figsize=(x,y))
```

```python
plt.xlabel('Total population')
plt.ylabel('Row Count')
plt.title('Histogram of population')
data['population'].hist(figsize=(10,5),color="pink",ec="black",lw=3) #hist(figsize=(x,y))
```

```python
plt.xlabel('households')
plt.ylabel('Row Count')
plt.title('Histogram of households')
data['households'].hist(bins=50,figsize=(10,5),color = "lightblue", ec="red", lw=3 ) #hist(figsize=(x,y)) bins=50, color = "lightblue", ec="red", lw=5
```

```python
plt.xlabel('median_income')
plt.ylabel('Row Count')
plt.title('Histogram of median_income')
data['median_income'].hist(bins=50,figsize=(10,5),color = "lightblue", ec="black", lw=3 ) #hist(figsize=(x,y)) bins=50, color = "lightblue", ec="red", lw=5
```

```python
plt.xlabel('median_house_value')
plt.ylabel('Row Count')
plt.title('Histogram of median_house_value')
data['median_house_value'].hist(bins=50,figsize=(10,5),color = "lightblue", ec="orange", lw=3 ) #hist(figsize=(x,y)) bins=50, color = "lightblue", ec="red", lw=5
```

```python
sns.set()
plt.figure(figsize=(10,8)) #Figure size
plt.scatter('longitude','latitude', data = data)
plt.ylabel('Latitudes')
plt.xlabel('Longitudes')
plt.title('Geographical plot of Lats/Lons')
plt.show()
```

```python
sns.set()
plt.figure(figsize=(10,8))#Figure size
plt.scatter('longitude','latitude',data=data,alpha=0.1)
plt.ylabel('Latitudes')
plt.xlabel('Longitudes')
plt.title('Geographical plot of Lats/Lons')
plt.show()
```

```python
plt.figure(figsize=(10,7))
data.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
        s=data["population"]/100, label="population", figsize=(15,8),
        c="median_house_value", cmap=plt.get_cmap("jet"),colorbar=True,
    )
plt.legend()
```

```python
plt.figure(figsize=(12,7))
plt.scatter('median_income','median_house_value',data=data,alpha=0.1)
plt.xlabel('Median income')
plt.ylabel('Median house value')
plt.title('Linear correlation Median income/Median House value')
```

```python
plt.subplot()
plt.title('Distribution Of Median age of a house within a block; a lower number is a newer building',fontsize=20)
sns.kdeplot(data['housing_median_age'],color='teal')
plt.show()
```

```python
plt.subplot()
plt.title('Distribution Of Median house value for households within a block (measured in US Dollars)',fontsize=20)
sns.kdeplot(data['median_house_value'],color='teal')
plt.show()
```

```python
plt.subplot()
plt.title('Distribution Of Total number of households, a group of people residing within a home unit, for a block',fontsize=20)
sns.kdeplot(data['households'],color='green')
plt.show()
```

```python
plt.subplot()
plt.title('Normalized Distribution Of The Total number of people residing within a block',fontsize=20)
data['population'] =np.log(data['population'])
sns.kdeplot(data['population'],color='green')
plt.show()
```

```python
plt.subplot()
plt.title('Normalized Distribution Of Total number of households, a group of people residing within a home unit, for a block',fontsize=20)
data['households'] =np.log(data['households'])
sns.kdeplot(data['households'],color='green')
plt.show()
```

```python
plt.subplot()
plt.title('Normalized Distribution Of Room Numbers',fontsize=20)
data['total_rooms'] =np.log(data['total_rooms'])
sns.kdeplot(data['total_rooms'])
plt.show()
```

```python
plt.subplot()
plt.title('Normalized Distribution Of Total Bedrooms',fontsize=20)
data['total_bedrooms'] =np.log(data['total_bedrooms'])
sns.kdeplot(data['total_bedrooms'])
plt.show()
```
